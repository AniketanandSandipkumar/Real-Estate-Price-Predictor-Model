```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('/kaggle/input/traindata/train.csv')
df
```

```python
df_=pd.read_csv('/kaggle/input/datatest/test.csv')
df_
```

```python
df2=df.merge(df_)
df2
```

```python
df2.drop('BHK_OR_RK',axis=1,inplace=True)
```

```python
df2['POSTED_BY'].unique()
```

```python
df_room=pd.get_dummies(df2['POSTED_BY'],drop_first=True)
df_room
```

```python
df4=pd.concat([df2,df_room],axis=1)
df4.head()
```

```python
df4.isnull().sum()
```

```python
df4 = df4.replace({True:1,False:0})
df4.drop('POSTED_BY',axis=1,inplace=True)
df4.head()
```

```python
df4['ADDRESS'].value_counts()
```

```python
df4.drop('ADDRESS',axis=1,inplace=True)
```

```python
df4.head()
```

```python
  correlation_matrix = df4.corr()
  print(correlation_matrix)
```

```python
df5=df4[['TARGET(PRICE_IN_LACS)']]
```

```python
df_train=df4[~df4['TARGET(PRICE_IN_LACS)'].isnull()]
df_train.head()
```

```python
X=df_train.drop('TARGET(PRICE_IN_LACS)',axis=1)
X.head()
```

```python
y=df_train[['TARGET(PRICE_IN_LACS)']]
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
from sklearn.ensemble import RandomForestRegressor
randomreg=RandomForestRegressor(n_estimators=50)
randomreg.fit(X_train,y_train)
```

```python
randomreg.score(X_test,y_test)
```

```python
y_predicted=randomreg.predict(X_test)
y_predicted
```

```python
y_test
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_predicted, alpha=0.5)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.show()
```
